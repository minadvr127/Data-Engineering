Time series data is said to be stationary if it does not exhibit trends or seasonality. That is, the mean, variance and covariance should be the same for any segment of the series, and are not functions of time.

Non-stationary data can be made to look stationary through differencing. A simple method called first order differencing calculates the difference between consecutive observations.

     𝑦′𝑡=𝑦𝑡−𝑦𝑡−1 
In this way a linear trend is transformed into a horizontal set of values.

```python
global_data['d1avg_temp'] = global_data['avg_temp'] - global_data['avg_temp'].shift(1)

global_data[['avg_temp','d1avg_temp']].head()

global_data['d1avg_temp'].plot(title="FIRST ORDER DIFFERENCE").autoscale(axis='x',tight=True);
```

as it can be seen from the above picture the temprature is statioanary

# What does the overall trend look like? Is the world getting hotter or cooler? Has the trend been consistent over the last few hundred years?

```python
plt.figure(figsize=(12, 5))

plt.plot(global_data['year'], global_data['avg_temp'], color = 'red', linewidth = 1.0 )

plt.show()
```

As you see the plot has some fluctiation but still a trend we can use Hodrick-Prescott filter which separates a time-series  𝑦𝑡  into a trend component  𝜏𝑡  and a cyclical component  𝑐𝑡 
𝑦𝑡=𝜏𝑡+𝑐𝑡 
The components are determined by minimizing the following quadratic loss function, where  𝜆  is a smoothing parameter:

min𝜏𝑡∑𝑇𝑡=1𝑐2𝑡+𝜆∑𝑇𝑡=1[(𝜏𝑡−𝜏𝑡−1)−(𝜏𝑡−1−𝜏𝑡−2)]2 
The  𝜆  value above handles variations in the growth rate of the trend component.
When analyzing quarterly data, the default lambda value of 1600 is recommended. Use 6.25 for annual data, and 129,600 for monthly data.

```python
from statsmodels.tsa.filters.hp_filter import hpfilter

# Tuple unpacking
gdp_cycle, gdp_trend = hpfilter(global_data['avg_temp'], lamb=1600)
```

```python
type(gdp_trend)
```

```python
global_data['trend'] = gdp_trend
global_data['cycle'] = gdp_cycle
```

```python
global_data[['trend','avg_temp']].plot().autoscale(axis='x',tight=True);
```

```python
Los_ange_cycle, Los_ange_trend = hpfilter(Los_ange_data['avg_temp'], lamb=1600)
```

```python
Los_ange_data['trend'] = Los_ange_trend
Los_ange_data['cycle'] = Los_ange_cycle
```

```python
Los_ange_data[['trend','avg_temp']].plot().autoscale(axis='x',tight=True);
```
